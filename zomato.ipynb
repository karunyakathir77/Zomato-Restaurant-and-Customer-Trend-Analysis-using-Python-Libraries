```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from IPython.display import display
```

```python
data = pd.read_csv('zomato.csv')
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
data.head()
```

```python
data.shape
```

```python
data.columns
```

```python
data.drop(['url','address','phone','menu_item','dish_liked','reviews_list'],axis=1,inplace=True)
data.head()
```

```python
data.isna().sum()
```

```python
data.shape
```

```python
data.drop_duplicates(inplace=True)
data.shape
```

```python
data['rate'].unique()
```

```python
data['rate'] = data['rate'].replace(['NEW', '-', ''], np.nan)
```

```python
data['rate'] = data['rate'].str.replace('/5','',regex=False).astype(float)
```

```python
data['rate'].head()
```

```python
data.rate.isna().sum()
```

```python
data.isna().sum()
```

```python
data.dropna(subset=['rate'], inplace=True)
```

```python
data.info()
```

```python
data.rename(columns={'approx_cost(for two people)':'Cost_2P', 'listed_in(type)':'Type'},inplace=True)
```

```python
data['Cost_2P'].unique()
```

```python
data['Cost_2P'] = data['Cost_2P'].str.replace(',','',regex=False).astype('Int32')
```

```python
rt_counts = data['rest_type'].value_counts()

rt_keep = rt_counts[rt_counts >= 500].index #to get the names/labels of rest

data['rest_type'] = data['rest_type'].apply(lambda x: x if x in rt_keep else 'Others') 
#it keeps if value in rest_type is same as rt_keep, else replace it with others
```

```python
cusinies_counts = data['cuisines'].value_counts()

cusinies_keep = cusinies_counts[cusinies_counts>=500].index

data['cuisines'] = data['cuisines'].apply(lambda x: x if x in cusinies_keep else 'Others')
```

```python
plt.style.use('ggplot')
```

```python
rest_counts = data['rest_type'].value_counts()

colors = ['skyblue' if rest == 'Others' else 'blue' for rest in rest_counts.index]

plt.figure(figsize=(10,6))
rest_counts.plot(kind='barh', color=colors)
plt.xlabel('Counts')
plt.ylabel('Types')
plt.title('Restaurant Types')
plt.grid(axis='x', linestyle='--')
plt.show()
```

```python
cuisine_cts = data['cuisines'].value_counts()

cuisine_cts = cuisine_cts[cuisine_cts.index != 'Others']

#colors_cusinies = ['skyblue' if cuisine == 'Others' else 'blue' for cuisine in cusinie_cts.index]

plt.figure(figsize=(10,6))
cuisine_cts.plot(kind='bar', color='seagreen')
plt.title('Popular Cuisines (Exclusing Others)')
plt.xlabel('Counts')
plt.ylabel('Types')
plt.grid(axis='x')
plt.show()
```

```python
location_counts = data['location'].value_counts().head(10)

location_counts.sort_values().plot(kind='barh', figsize=(10,6), color='darkblue')
plt.title('Top 10 Locations with Highest Number of Restaurants')
plt.xlabel('Restaurant Counts')
plt.ylabel('Locations')
```

```python
online_counts = data['online_order'].value_counts()

plt.figure(figsize=(5,5))
plt.pie(online_counts, labels=online_counts.index, 
        autopct='%1.1f%%', colors=['darkblue','skyblue'],
        wedgeprops={'edgecolor':'black'})
plt.legend()
plt.title('Online Order Availability')
```

```python
book_counts = data['book_table'].value_counts()

explode = [0.1 if label == 'Yes' else 0 for label in book_counts.index]

plt.figure(figsize=(5,5))
plt.pie(book_counts, labels=book_counts.index,
        autopct='%1.1f%%', colors=['darkblue','skyblue'], explode=explode,
        wedgeprops={'edgecolor':'black'}, startangle=140)
plt.title('Table Booking Availability')
plt.legend()
```

```python
votes_per_location = data.groupby('location')['votes'].sum().sort_values(ascending=False)

votes_per_location.head(15).plot(kind='bar', figsize=(10,6), color='coral')
plt.title('Total Votes by Top 15 Location')
plt.xlabel('Location')
plt.ylabel('Total Votes')
plt.tight_layout()
```

```python
avg_rating_by_location = data.groupby('location')['rate'].mean().sort_values(ascending=False)

high_rating_location = avg_rating_by_location[avg_rating_by_location >= 4.0]

high_rating_location.plot(kind='bar', figsize=(10,6), color='teal')
plt.title('Locations with Average Rating >= 4.0')
plt.xlabel('Location')
plt.ylabel('Average Rating')
plt.xticks(rotation=0)
plt.ylim(4,4.5)
```

```python
high_rated_restaurants = data[data['rate'] >= 4.0]

high_rated_restaurants[['name','location','rate']].drop_duplicates().sort_values(by='rate', ascending=False).head(10)
```

```python
high_cost_2P = data[data['Cost_2P'] >= 2000]

high_cost_2P[['name','location','Cost_2P','rate']].drop_duplicates().sort_values(by='Cost_2P',ascending=False).head(5)
```
